## 데이터 읽어오기

```
import pandas as pd

src = pd.read_csv('https://s3.ap-northeast-2.amazonaws.com/data10902/petition/petition.csv')
```

데이터를 읽어오려면 오래 걸리니까 위에서 한 번만 읽어와서 `src`에 저장하고, 이후부터는 복사본을 `df`에 저장하여 쓴다. 하다가 뭔가 잘못되면 `src`를 다시 복사하면 되니까 효율적.

```
# 복사본 만들기
df = src.copy()

# 결측치가 있는 행은 제거
df = df.dropna()
```

```
single_row = pd.DataFrame({"번호":1, "분류":'b', \
                             "제목":'c', "참여인원":0}, index=[0])
single_row
```

```
#학교가나오거나안나오거나, 숫자123중에하나가 나온다
df.loc[df['content'].str.contains(r'(중|고등)(학교)?\s?[123]', regex=True)]  
```

```
df['content'].str.extractall(r'([\d,]+원)')
```

```
import re
re.sub('\\D', '', "정가 15,300원")
```

```
import re
int(re.sub('\\D', '', "정가 15,300원"))   # \d\d 는 숫자2개를 뜻함
```

```
# raw string 리터럴
r"\d"   

'(중|고등)(학교)?\\s?[123]' → r'(중|고등)(학교)?\s?[123]'
'\\d+원' → r'\d+원'
'\\D' → r'\D'
```

```
df['content'].str.extractall(r'(0\d\d[-\.]\d\d\d\d[-\.]\d\d\d\d)')
```

```
df['content'].str.extractall(r'(0\d\d-\d\d\d\d-\d\d\d\d)')
```

```
df['content'].str.extractall(r'(\d+원을.+손해)')
```

```
df['content'].str.extractall(r'([김이박정차]\S\S[씨님])')   # 중1~3이나 고1~3을 뽑아냄
```

```
df['content'].str.extractall(r'((존경|친애|경애)하옵?는\s\S+[님께])') 
```
